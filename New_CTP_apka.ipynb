```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('GL02.csv')
```

```python
print(df)
```

```python
dataframe = pd.DataFrame(df)
```

```python
print(dataframe)
```

```python
newdf = dataframe[dataframe['X_Value'] %5.0 == 0.0]
```

```python
print(newdf)
```

```python
plt.figure(num=None, figsize=(15, 10), dpi=240)
plt.plot(newdf["X_Value"], newdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newdf["X_Value"], newdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newdf["X_Value"], newdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newdf["X_Value"], newdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

x = newdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.ylim(-10.00, 10.00)

plt.title('Wykres napięcia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend()
plt.savefig('Wykres1.pdf')
plt.show()
```

```python
newSdf = dataframe[dataframe['X_Value'] %1.0 == 0.0]
```

```python
print(newSdf)
```

Ponowny wydruk wykresu - widok od 70 do 80 sekundy co sekundę

```python
plt.figure(num=None, figsize=(15, 10), dpi=240)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

x = newdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.xlim(70.00, 80.00)
plt.ylim(-10.00, 10.00)

plt.title('Wykres napięcia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend()
plt.savefig('Wykres1-zblizenie.pdf')
plt.show()
```
