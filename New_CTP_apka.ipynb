```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('GL02.csv')
```

```python
dataframe = pd.DataFrame(df)
```

```python
print(dataframe)
```

```python
newSdf = dataframe[dataframe['X_Value'] %0.5 == 0.0]
```

```python
print(newSdf)
```

Wydruk wykresu - widok z danymi co 0,5 sekundy

```python
plt.figure(num=None, figsize=(15, 10), dpi=120)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

x = newSdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.ylim(-10.00, 10.00)

plt.title('Wykres napięcia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend()
plt.savefig('Wykres1.pdf')
plt.show()
```

Ponowny wydruk wykresu ze zbliżeniem na 30-sekundowy okres między 60 a 90 sekundą

```python
plt.figure(num=None, figsize=(15, 10), dpi=120)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

x = newSdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.xlim(60.00, 90.00)
plt.ylim(-10.00, 10.00)

plt.title('Wykres napięcia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend()
plt.savefig('Wykres1-zblizenie.pdf')
plt.show()
```

Wydruk wykresu z zaznaczonymi wartościami maks., min., oraz średnią dla każdej krzywej napięcia

```python
plt.figure(num=None, figsize=(15, 10), dpi=120)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

plt.axhline(float(mean0), c = 'red', label = 'mean V0', linestyle='dashed', linewidth=2)
plt.axhline(float(mean1), c = 'purple', label = 'mean V1', linestyle='dashed', linewidth=2)
plt.axhline(float(mean2), c = 'orange', label = 'mean V2', linestyle='dashed', linewidth=2)
plt.axhline(float(mean3), c = 'turquoise', label = 'mean V3', linestyle='dashed', linewidth=2)

plt.scatter(1, max0, c = 'red', label = 'max V0', marker = '^', s = 10**2)
plt.scatter(1, min0, c = 'red', label = 'min V0', marker = 'v', s = 10**2)
plt.scatter(3, max1, c = 'purple', label = 'max V1', marker = '^', s = 10**2)
plt.scatter(3, min1, c = 'purple', label = 'min V1', marker = 'v', s = 10**2)
plt.scatter(5, max2, c = 'orange', label = 'max V2', marker = '^', s = 10**2)
plt.scatter(5, min2, c = 'orange', label = 'min V2', marker = 'v', s = 10**2)
plt.scatter(7, max3, c = 'turquoise', label = 'max V3', marker = '^', s = 10**2)
plt.scatter(7, min3, c = 'turquoise', label = 'min V3', marker = 'v', s = 10**2)

x = newSdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.ylim(-10.00, 10.00)

plt.title('Wykres napięcia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.savefig('Wykres1_maxmin.pdf')
plt.show()
```
