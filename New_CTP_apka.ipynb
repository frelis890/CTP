```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('GL02.csv')
```

```python
dataframe = pd.DataFrame(df)
```

```python
print(dataframe)
```

```python
newSdf = dataframe[dataframe['X_Value'] %0.5 == 0.0]
```

```python
print(newSdf)
```

Wydruk wykresu - widok z danymi co 0,5 sekundy

```python
plt.figure(num=None, figsize=(15, 10), dpi=120)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

x = newSdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.ylim(-10.00, 10.00)

plt.title('Wykres napiÄ™cia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend()
plt.savefig('Wykres1.pdf')
plt.show()
```

Ponowny wydruk wykresu ze zbliÅ¼eniem na 30-sekundowy okres miÄ™dzy 60 a 90 sekundÄ…

```python
plt.figure(num=None, figsize=(15, 10), dpi=120)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')

x = newSdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.xlim(60.00, 90.00)
plt.ylim(-10.00, 10.00)

plt.title('Wykres napiÄ™cia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend()
plt.savefig('Wykres1-zblizenie.pdf')
plt.show()
```

Wydruk wykresu z zaznaczonymi wartoÅ›ciami maks., min., oraz Å›redniÄ… dla kaÅ¼dej krzywej napiÄ™cia

```python
plt.figure(num=None, figsize=(15, 10), dpi=120)
plt.plot(newSdf["X_Value"], newSdf["Voltage_0"], c = 'red', label = 'Voltage 0')
plt.plot(max0, c = 'red', label = 'max V0', marker = '^', markersize = 15)
plt.plot(min0, c = 'red', label = 'min V0', marker = 'v', markersize = 15)
plt.plot(mean0, c = 'red', label = 'mean V0', marker = 'o', markersize = 15)

plt.plot(newSdf["X_Value"], newSdf["Voltage_1"], c = 'purple', label = 'Voltage 1')
plt.plot(max1, c = 'purple', label = 'max V1', marker = '^', markersize = 15)
plt.plot(min1, c = 'purple', label = 'min V1', marker = 'v', markersize = 15)
plt.plot(mean1, c = 'purple', label = 'mean V1', marker = 'o', markersize = 15)

plt.plot(newSdf["X_Value"], newSdf["Voltage_2"], c = 'orange', label = 'Voltage 2')
plt.plot(max2, c = 'orange', label = 'max V2', marker = '^', markersize = 15)
plt.plot(min2, c = 'orange', label = 'min V2', marker = 'v', markersize = 15)
plt.plot(mean2, c = 'orange', label = 'mean V2', marker = 'o', markersize = 15)

plt.plot(newSdf["X_Value"], newSdf["Voltage_3"], c = 'turquoise', label = 'Voltage 3')
plt.plot(max3, c = 'turquoise', label = 'max V3', marker = '^', markersize = 15)
plt.plot(min3, c = 'turquoise', label = 'min V3', marker = 'v', markersize = 15)
plt.plot(mean3, c = 'turquoise', label = 'mean V3', marker = 'o', markersize = 15)

x = newSdf["X_Value"]
plt.xticks(np.arange(min(x), max(x)+1, 5.0))
plt.ylim(-10.00, 10.00)

plt.title('Wykres napiÄ™cia w czasie')
plt.xlabel('Time[s]')
plt.ylabel('Voltage')

plt.axhline(0, color='black')  # x = 0
plt.axvline(0, color='black')  # y = 0

plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.savefig('Wykres1_maxmin.pdf')
plt.show()
```
